```python
import csv
import pandas as pd
from plotnine import *
from matplotlib import pyplot as plt
from tqdm import tqdm

data = []
with open('../cldf/forms.csv') as fin:
    reader = csv.reader(fin)
    next(reader)
    for x in tqdm(reader):
        if x[3] and x[1] in ['K', 'Indo-Aryan', 'Pa']: data.append([x[1], len(x[3])])

df = pd.DataFrame(data, columns=['Lang', 'Len'])
```

```python
df.head()
```

```python
ggplot(df, aes(x='Len', fill='Lang')) + geom_histogram(binwidth=1) + facet_wrap('Lang', scales='free_y')
```
